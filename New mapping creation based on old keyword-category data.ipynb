```python
import pandas as pd
```

```python
data = pd.read_csv("bleaches_purplle.csv") #Upload the old category file downloaded from Athena
data.head(20)
```

```python
print(data.isnull().sum()) #Checking the null values
```

```python
print(data.fillna(value="", inplace=True)) #Fill null values by blanks
```

```python
print(data.shape)
```

```python
#Remove the duplicate rows by keeping top score mapping
data.drop_duplicates(subset=["keyword", "actual_keyword", "category_l1", "category_l2", "category_l3"], keep='first', inplace=True)
data.reset_index(drop=True, inplace=True)
print(data.shape)
```

```python
print(data["category_l3"].unique())
```

```python
data["new_category_l2"] = "facial kit & bleach" #adding extra columns with new caytegory l2
data["new_source"] = "manual"
data.loc[data['category_l3'] == 'bleaches', 'new_category_l3'] = 'bleach' #Appending a new category_l3 in an additional column when the old category_l3 matches the existing category that needs to be updated.
```

```python
print(data.head())
```

```python
#Exclude category L2 or L3 for which a new column has been generated.
data.drop(["category_l2", "category_l3", "source"], axis=1, inplace=True)
# data.drop(["category_l3", "source"], axis=1, inplace=True)
```

```python
#Rename the columns
data.rename(columns={"new_category_l2":"category_l2", "new_category_l3":"category_l3", "new_source":"source"}, inplace=True)
```

```python
print(data.head())
```

```python
#Converting data into correct format
data_with_new_categories = data[['keyword', 'actual_keyword', 'category_l1', 'category_l2', 'category_l3', 'category_l4', 'category_l5', 'category_l6',
       'category_l7', 'category_l8', 'count', 'score', 'source']]
print(data_with_new_categories.head())
```

```python
print(data_with_new_categories["category_l3"].unique())
```

```python
#Saving the data in CSV format
data_with_new_categories.to_csv("bleach_category_mapping_purplle_11thJan(CSV).csv" , index=False)
```
